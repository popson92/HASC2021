```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data_energy = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/00374/energydata_complete.csv')
```

## renaming some columns

```python
columns_names = { "T1": "Temperature_kitchen","RH_1": "Humidity_kitchen", "T2": "Temperature_living",
                 "RH_2": "Humidity_living", "T3": "Temperature_laundry", "RH_3": "Humidity_laundry",
                 "T4": "Temperature_office", "RH_4": "Humidity_office", "T5":"Temperature_bathroom",
                 "RH_5": "Humidity_bathroom", "T6": "Temperature_outside", "RH_6": "Humidity_outside",
                 "T7": "Temperature_ironing", "RH_7": "Humidity_ironing", "T8": "Temperature_teenager",
                 "RH_8": "Humidity_teenager", "T9": "Temperature_parents", "RH_9": "Humidity_parents", 
                 "To": "Temperature_outside", "rv1": "Random_variable1", "rv2": "Random_variable2" }
```

```python
data_energy = data_energy.rename(columns=columns_names)
```

```python
data_energy.columns
```

```python
data_energy.info()
```

From the above cell, we can see that the most of the columns of the dataset are of numeric data types, Only the date is an object data type and from instructions it is not needed for the task of the quiz

### Dropping the two columns from instruction

```python
data_energy = data_energy.drop(['date', 'lights'], axis = 1) 
data_energy.head(10).T
```

#### Performing normalisation and splitting the data

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error

scaler = MinMaxScaler()
normalised_data_energy = pd.DataFrame(scaler.fit_transform(data_energy), columns=data_energy.columns)
features = normalised_data_energy.drop(columns=['Appliances'])
target = normalised_data_energy['Appliances']
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)

print('X_train shape: {}'.format(X_train.shape))
print('y_train shape: {}'.format(y_train.shape))
print('X_test shape: {}'.format(X_test.shape))
print('y_test shape: {}'.format(y_test.shape))
```

##### Working on T2 and T6

```python
df_reg = normalised_data_energy[["Temperature_living", "Temperature_outside"]]
df_reg.head().T
```

#### Splitting and reshaping "df_reg' into train and test set

```python
x= df_reg["Temperature_living"].values.reshape(-1,1)
y = df_reg["Temperature_outside"].values.reshape(-1,1)

xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
small_regr = LinearRegression()

# Train the model using the training sets
small_regr.fit(xtrain, ytrain)

# Make predictions using the testing set
pred = small_regr.predict(xtest)
```

```python
pred
```

```python
from sklearn.metrics import r2_score

r2_score = r2_score(ytest, pred)
print('R-squared:',(round(r2_score, 2)))
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(ytest, pred)

print('MAE:',(round(mae, 2)))
```

```python
rss = np.sum(np.square(ytest - pred))
print('RSS:',(round(rss, 2)))
```

```python
from sklearn.metrics import  mean_squared_error
rmse = np.sqrt(mean_squared_error(ytest, pred))
print('RMSE:',(round(rmse, 3)))
```

### Working on the Main dataset

```python
regr = LinearRegression()

# Train the model using the training sets
regr.fit(X_train, y_train)

# Make predictions using the testing set
energy_pred = regr.predict(X_test)

print("Training set score: {:.3f}".format(regr.score(X_train, y_train)))
print("Test set score: {:.3f}".format(regr.score(X_test, y_test)))
```

```python
def get_weights_df(model, feat, col_name):
  #this function returns the weight of every feature
  weights = pd.Series(model.coef_, feat.columns).sort_values()
  weights_df = pd.DataFrame(weights).reset_index()
  weights_df.columns = ['Features', col_name]
  weights_df[col_name].round(3)
  return weights_df
```
